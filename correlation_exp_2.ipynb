```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('House_Rent_Dataset.csv')
df1
```

```python
df1.head()
```

```python
df1.tail()
```

```python
corr = df1.corr()
corr

```

```python
plt.figure(figsize=(5,2), dpi =200)
sns.heatmap(corr,annot=True,fmt=".2f", linewidth=.5)
plt.show()
```
